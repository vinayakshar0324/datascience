```python
#import all major libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#usage of inline plotting
%matplotlib inline
```

```python
#load data from CSV File
dataFrame = pd.read_csv('./911.csv')
```

```python
dataFrame.info
```

```python
#check head for our dataframe
dataFrame.head()
```

```python
#top 5 zip codes
dataFrame['zip'].value_counts().head(5)
```

```python
#top 5 twonships
dataFrame['twp'].value_counts().head(5)
```

```python
#unique reasons to call 911
dataFrame['title'].nunique()
```

```python
#Top 5 specific reasons to call 911
#Assignment was there
dataFrame['SpecificReason'] = dataFrame['title'].apply(lambda title: title.split(':')[0])
```

```python
dataFrame.head()
```

```python
dataFrame['SpecificReason'].value_counts().head(20)
```

```python
#plot a countplt 
sns.countplot(x='SpecificReason', data=dataFrame)
```

```python
#convert time stamp from object to actual timestamp
dataFrame['timeStamp'] = pd.to_datetime(dataFrame['timeStamp'])
```

```python
dataFrame.info
```

```python
type(dataFrame['timeStamp'].iloc[0])
```

```python
#create 3 coulms fro hours, months and days of the week
dataFrame['Hours'] = dataFrame['timeStamp'].apply(lambda time: time.hour)
```

```python
dataFrame.head()
```

```python
dataFrame['Hours'] = dataFrame['timeStamp'].apply(lambda time: time.hour)
dataFrame['Month'] = dataFrame['timeStamp'].apply(lambda time: time.month)
dataFrame['Days'] = dataFrame['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dataFrame.head(20)
```

```python
dataFrame['Days'].nunique()
```

```python
#plot a graph for month 
sns.countplot(x='Month', data=dataFrame, hue='SpecificReason')
```

```python
#use group_by

byMonth = dataFrame.groupby('Days').count()
```
